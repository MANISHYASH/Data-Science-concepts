Linear Regression
-------------------

y = mx + b

where .,

m = slope

b = intercept

y = dependant variable

x = indepedant variable

To find m:

Calculate the slope (𝑚) as the rise-over-run between two points on the line:

𝑚 = Δ𝑦/Δ𝑥 = change in price/change in area


​


For example:

From your dataset, choose two points on the line (e.g., (2600, 550000) and (3000, 565000)).

Compute:

𝑚 = (565000 − 550000) / (3000 − 2600) ==> 15000 / 400 = 39.25

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Create the DataFrame
data = {
    "area": [2600, 3000, 3200, 3600, 4000],
    "price": [550000, 565000, 610000, 680000, 725000]
}
df = pd.DataFrame(data)

# Initialize the Linear Regression model
model = LinearRegression()

# Train the model (fit)
model.fit(df[['area']], df['price'])

# Print the coefficients
print("Coefficient (slope):", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Create the DataFrame
data = {
    "area": [2600, 3000, 3200, 3600, 4000],
    "price": [550000, 565000, 610000, 680000, 725000]
}
df = pd.DataFrame(data)

# Train the Linear Regression model
model = LinearRegression()
model.fit(df[['area']], df['price'])

# Predict prices for the regression line
predicted_prices = model.predict(df[['area']])

# Scatter plot of the data
plt.scatter(df['area'], df['price'], color='red', marker='+', label='Actual Price')

# Plot the regression line
plt.plot(df['area'], predicted_prices, color='blue', label='Regression Line')

# Labels and title
plt.xlabel('Area (square feet)')
plt.ylabel('Price (in dollars)')
plt.title('Scatter Plot with Regression Line: Area vs Price')

# Show legend
plt.legend()

# Show the plot
plt.show()
```
